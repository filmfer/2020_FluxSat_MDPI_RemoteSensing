```python
# Importar bibliotecas
import pandas as pd
import numpy as np
import matplotlib as mp

# Funções

## ========== ***** ========== ##

# Caminho/link para base de daos no Google Docs (GSheets):
#   Alterar:  "edit?gid=1310164497#gid=1310164497"
#   Para:     "export?format=csv&gid=1310164497"
# Identificação da folha: gid=1310164497 (Muda de folha para folha e de documento para decomento)
# Exportat tipo de ficheiro: export?format=csv (Pode ser: CSV, XLSX ou TSV)

#path ="https://docs.google.com/spreadsheets/d/1K7WgxMUH2nlWCT4JH0T8tZZT6_6Gu4ur0-Kx3eobHKI/export?format=csv&gid=1310164497"
#path = "https://docs.google.com/spreadsheets/d/1YIT8OFE5KstW7sq0VG4zbuPELACM96Gb/export?format=csv&gid=1955173554"
path = "https://docs.google.com/spreadsheets/d/1YIT8OFE5KstW7sq0VG4zbuPELACM96Gb/export?format=csv&gid=561767035"

# Carregar dados
data = pd.read_csv(path)
#Definição do "Framework" para o Pandas trabalhar os dados
df = pd.DataFrame(data)
#print(df)

# Calcular totais por coluna/variável
# Calcular total de Peso bruto (g)
print("Peso bruto (g)")
df["Pesobruto(g)"] = pd.to_numeric(df["Pesobruto(g)"], errors='coerce')
seeds = df["Pesobruto(g)"].sum()
#seeds = df["Pesobruto(g)"].sum(skipna=True), errors='coerce')
print("Total sementes recolhidas: " + str((f"{seeds:.2f}")) + " g")
print("Total sementes recolhidas: " + str((f"{(seeds/1000):.2f}")) + " kg.")
# Calcular total de Peso neto (g)
print("Peso neto (g)")
seedsn = df["Pesoneto(g)"].sum(skipna=True)
print("Total sementes recolhidas: " + str((f"{seedsn:.2f}")) + " g")
print("Total sementes recolhidas: " + str((f"{(seedsn/1000):.2f}")) + " kg.")

# Criar lista de valores únicos na colunas
ilhas = df['Ilha'].unique().tolist()
print(ilhas)
locais = df['Local'].unique().tolist()
species = df['Espécie'].unique().tolist()
ano = df['Ano'].unique().tolist()

# Criar lista de valores únicos na coluna por ilha
for i in ilhas:
    isp = df.groupby('Ilha')['Espécie'].unique().tolist()

# Confirmar # de ilhas, não pode ser superior a 9
if len(ilhas) > 9:
    print("Há mais de 9 ilhas! Verificar nomes na tabela original.")
    print(ilhas)
else:
    print("Número de ilhas correcto: " + str(len(ilhas)))

print("Recolha de sementes em: " + str(len(locais)) + " locais.")
#print(locais)
print("Recolha de sementes das espécies: " + str(len(species)) + " espécies distintas.")
#print(species)

print("")
print(" ========== ***** ========== ")
print("Total de sementes recolhidas (peso bruto)  por ilha:")
for i in ilhas:
    df["Pesobruto(g)"] = pd.to_numeric(df["Pesobruto(g)"], errors='coerce')
    ilha = df.groupby('Ilha')['Pesobruto(g)'].sum()
    print(i + ": " + str((f"{ilha[i]:.2f}")))

print("")
print(" ========== ***** ========== ")
print("Total de sementes recolhidas (peso neto)  por ilha:")
for i in ilhas:
    ilha = df.groupby('Ilha')['Pesoneto(g)'].sum()
    print(i + ": " + str((f"{ilha[i]:.2f}")))

print("")
print(" ========== ***** ========== ")
print("Total sementes recolhidas (peso bruto) por espécie:")
for s in species:
    df["Pesobruto(g)"] = pd.to_numeric(df["Pesobruto(g)"], errors='coerce')
    especie = df.groupby('Espécie')['Pesobruto(g)'].sum()
    print(s + ": " + str((f"{especie[s]:.2f}")))

print("")
print(" ========== ***** ========== ")
print("Total sementes recolhidas (peso neto) por espécie:")
for s in species:
    especie = df.groupby('Espécie')['Pesoneto(g)'].sum()
    print(s + ": " + str((f"{especie[s]:.2f}")))


print("")
print(" ========== ***** ========== ")
print("Total sementes recolhidas (peso bruto) por ilha e por espécie:")

for a in ano:
  for i in ilhas:
    for s in species:
      filtro_isp = df[(df['Ano'] == a) & (df['Ilha'] == i) & (df['Espécie'] == s)]
      df["Pesobruto(g)"] = pd.to_numeric(df["Pesobruto(g)"], errors='coerce')
      isp = filtro_isp['Pesobruto(g)'].sum(skipna=True)
      if isp == 0:
          continue
      else:
        print(str(a) + " - " + str(i) + " - " + str(s) + ": " + str((f"{isp:.2f}")))

print("")
print(" ========== ***** ========== ")
print("Total sementes recolhidas (peso neto) por ilha e por espécie:")

for a in ano:
  for i in ilhas:
    for s in species:
      filtro_isp = df[(df['Ano'] == a) & (df['Ilha'] == i) & (df['Espécie'] == s)]
      isp = filtro_isp['Pesoneto(g)'].sum(skipna=True)
      if isp == 0:
          continue
      else:
        print(str(a) + " - " + str(i) + " - " + str(s) + ": " + str((f"{isp:.2f}")))

print("")
print(" ========== ***** ========== ")
```
