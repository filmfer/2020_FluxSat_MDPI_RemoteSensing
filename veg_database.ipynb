```python
# Importar bibliotecas
import pandas as pd
import numpy as np
import matplotlib as mp

# Funções

## ========== ***** ========== ##

# Caminho/link para base de daos no Google Docs (GSheets):
#   Alterar:  "edit?gid=1310164497#gid=1310164497"
#   Para:     "export?format=csv&gid=1310164497"
# Identificação da folha: gid=1310164497 (Muda de folha para folha e de documento para decomento)
# Exportat tipo de ficheiro: export?format=csv (Pode ser: CSV, XLSX ou TSV)

#path ="https://docs.google.com/spreadsheets/d/1K7WgxMUH2nlWCT4JH0T8tZZT6_6Gu4ur0-Kx3eobHKI/export?format=csv&gid=1310164497"
path = "https://docs.google.com/spreadsheets/d/1YIT8OFE5KstW7sq0VG4zbuPELACM96Gb/export?format=csv&gid=2097175196"

# Carregar dados
data = pd.read_csv(path)
#Definição do "Framework" para o Pandas trabalhar os dados
df = pd.DataFrame(data)
#print(df)

# Calcular totais por coluna/variável
# Calcular total de # de individuos plantados
plantadas = df["# indivíduos plantados"].sum(skipna=True)
print("Total de arbóreas plantadas: " + str(int(plantadas)))

# Área total de plantações (estimativa)
areaplant = df["Tamanho área de plantação (m2)"].sum(skipna=True)
print("Área total (m^2) de arbóreas plantadas: " + str((f"{areaplant:.2f}")) + " m^2.")
print("Área total (m^2) de arbóreas plantadas: " + str((f"{areaplant/10000:.2f}")) + " ha.")

# Numero total de dias de plantação
diasp = df["# dias trabalhados na plantação"].sum(skipna=True)
print("Numero total de dias de plantação: " + str(diasp))

# Criar lista de valores únicos na colunas
ilhas = df['Ilha'].unique().tolist()
locais = df['Local'].unique().tolist()
species = df['Espécie plantada'].unique().tolist()

# Criar lista de valores únicos na coluna por ilha
for i in ilhas:
    isp = df.groupby('Ilha')['Espécie plantada'].unique().tolist()

# Confirmar # de ilhas, não pode ser superior a 9
if len(ilhas) > 9:
    print("Há mais de 9 ilhas! Verificar nomes na tabela original.")
    print(ilhas)
else:
    print("Número de ilhas correcto: " + str(len(ilhas)))

print("Plantação de arbóreas em: " + str(len(locais)) + " locais.")
#print(locais)
print("Plantação de espécies: " + str(len(species)) + " espécies distintas.")
#print(species)

print("")
print(" ========== ***** ========== ")
print("Total de dias de plantação por ilha:")
for i in ilhas:
    ilha = df.groupby('Ilha')['# dias trabalhados na plantação'].sum()
    print(i + ": " + str(int(ilha[i])))

print("")
print(" ========== ***** ========== ")
print("Total de indivíduos de arbóreas plantados por ilha:")
for i in ilhas:
    ilha = df.groupby('Ilha')['# indivíduos plantados'].sum()
    print(i + ": " + str(int(ilha[i])))

#for l in locais:
#    local = df.groupby('Local')['# indivíduos plantados'].sum()
#    print(l + ": " + str(int(local[l])))

print("")
print(" ========== ***** ========== ")
print("Total de indivíduos plantados por espécie:")
for s in species:
    especie = df.groupby('Espécie plantada')['# indivíduos plantados'].sum()
    print(s + ": " + str(int(especie[s])))

print("")
print(" ========== ***** ========== ")
print("Total de indivíduos plantados por ilha e por espécie:")

for i in ilhas:
  for s in species:
    #print(i, s)
    #isp = df.groupby('Espécie plantada')['# indivíduos plantados'].sum()
    #isp = df[(df['Ilha'] == i) & (df['Espécie plantada'] == s).sum()]
    filtro_isp = df[(df['Ilha'] == i) & (df['Espécie plantada'] == s)]
    isp = filtro_isp['# indivíduos plantados'].sum(skipna=True)
    if isp == 0:
        continue
    else:
      print(str(i) + " - " + str(s) + ": " + str(int(isp)))
print("")
print(" ========== ***** ========== ")
```
